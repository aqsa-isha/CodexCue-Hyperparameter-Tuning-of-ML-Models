```python
# Step 1: Identify non-numeric columns
non_numeric_cols = data.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)

# Step 2: Drop non-numeric columns
X_cleaned = data.drop(columns=non_numeric_cols)
```

#### Hyperparameters are configuration settings used to optimize the performance of machine learning models.
#### Unlike parameters learned from training data (e.g., weights in a neural network),
#### hyperparameters are set before the training process begins and determine the behavior of the training algorithm.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
X = data.drop(columns = ['Prediction'])
y = data['Prediction']
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(random_state = 42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs = 1, verbose = 2)
grid_search.fit(X_train, y_train)

print('Best parameters from Grid Search', grid_search.best_params_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
model = RandomForestClassifier(random_state = 42)
param_dist = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6]
}
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, cv=3, random_state=42, n_jobs = -1, verbose = 2)
random_search.fit(X_train, y_train)
print("Best Parameters for Random Search ", random_search.best_params_)
```

```python
from sklearn.metrics import accuracy_score, classification_report
best_grid_model = grid_search.best_estimator_
y_pred_grid = best_grid_model.predict(X_test)
print(f"Grid Search Model Accuracy: {accuracy_score(y_test, y_pred_grid)}")
print(classification_report(y_test, y_pred_grid))

best_random_model = random_search.best_estimator_
y_pred_random = best_random_model.predict(X_test)
print(f"Random Search Model Accuracy: {accuracy_score(y_test, y_pred_random)}")
print(classification_report(y_test, y_pred_random))
```

### Random Search is usually faster than Grid Search
